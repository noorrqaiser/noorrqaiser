MIMIC-IV-ED is a publicly accessible database of over 400,000 emergency department (ED) admissions to the Beth Israel Deaconess Medical Center between 2011 and 2019.

The MIMIC-IV-ED demo is a relational database **containing a single patient tracking table, edstays, and five data tables: diagnosis, medrecon, pyxis, triage, and vitalsign.** All tables from MIMIC-IV-ED were included in the demo dataset.


**Johnson, A., Bulgarelli, L., Pollard, T., Celi, L. A., Horng, S., & Mark, R. (2023). MIMIC-IV-ED Demo (version 2.2). PhysioNet. https://doi.org/10.13026/jzz5-vs76.**

```python
#Load the Data and Relevant Libraries
import pandas as pd
path = "/content/drive/MyDrive/mimicPRACTICE.csv"
df = pd.read_csv(path)
#Print the first 10 rows to view the data and columns
df.head(10)
```

```python
#Simple Descriptive Statistics
df.describe()
```

```python
#import more libraries
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
# Convert intime and outtime to datetime
df['intime'] = pd.to_datetime(df['intime'])
df['outtime'] = pd.to_datetime(df['outtime'])
```

```python
df['intime']
```

```python
# Calculate duration of stay
df['duration_of_stay'] = (df['outtime'] - df['intime']).dt.total_seconds() / 3600
print(df['duration_of_stay'])
```

DATA CLEANING

```python
df.isnull().sum()
```

```python
# Iterate over columns and fill missing values with mean only for numeric columns
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
      # Check if column is numeric
        df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Normalize the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df[['temperature', 'heartrate', 'resprate', 'o2sat', 'sbp', 'dbp', 'acuity', 'duration_of_stay']])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_df)
df['cluster'] = clusters
```

```python
# Evaluate clusters
silhouette_avg = silhouette_score(scaled_df, clusters)
print(f'Silhouette Score: {silhouette_avg}')
```

```python
# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature', y='heartrate', hue='cluster', data=df, palette='viridis')
plt.title('Clusters of Patients')
plt.show()
```

```python
#Predictive modeling
#Split the data into training and testing sets
#create input-data and target-data
x_data = df[['temperature', 'heartrate', 'resprate', 'o2sat', 'sbp', 'dbp', 'duration_of_stay']]
y_data = df['race']
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
# Applying a Logistic Regression on all attributes and report the performance for all metrics
# including accuracy, precision, recall, and AUC

from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, PolynomialFeatures


# Instantiate the LogisticRegression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(x_train, y_train)

# Make predictions on the test data
y_pred = model.predict(x_test)

# Calculate the accuracy, precision, recall, and AUC
accuracy = model.score(x_test, y_test)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
# Try other models for comparison
# For example: RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
rf_y_pred = rf_model.predict(x_test)
rf_y_pred_proba = rf_model.predict_proba(x_test)

# Calculate metrics for RandomForestClassifier
rf_accuracy = accuracy_score(y_test, rf_y_pred)
rf_precision = precision_score(y_test, rf_y_pred, average='macro')
rf_recall = recall_score(y_test, rf_y_pred, average='macro')

# Print the metrics for RandomForestClassifier
print("\nRandomForestClassifier Metrics")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
```
